```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("PRSA_data_2010.1.1-2014.12.31.csv")
```

```python
# Clean the data
data = data.drop_duplicates()
data = data.dropna(subset=['pm2.5', 'cbwd', 'month', 'year'])
```

```python
# Daily PM2.5 Levels
plt.figure(figsize=(10, 6))
plt.plot(data["pm2.5"])
plt.title("Daily PM2.5 Levels")
plt.xlabel("Time (Days)")
plt.ylabel("PM2.5 Levels (µg/m³)")
plt.grid(True)
plt.show()
```

```python
# Average PM2.5 Levels by Location and Month
pivot_table = data.groupby(['cbwd', 'month'])['pm2.5'].mean().unstack()

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap='YlGnBu', linewidths=.5)
plt.title('Average PM2.5 Levels Across Different Locations and Months')
plt.xlabel('Month')
plt.ylabel('Location (cbwd)')
plt.show()
```

```python
# Correlation Between PM2.5 Levels and Temperature
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TEMP', y='pm2.5', data=data, alpha=0.5)
plt.title('Correlation Between PM2.5 Levels and Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('PM2.5 Levels (µg/m³)')
plt.grid(True)
plt.show()
```

```python
# Distribution of PM2.5 Levels by Month
plt.figure(figsize=(12, 6))
sns.boxplot(x='month', y='pm2.5', data=data , palette='Accent')
plt.title('Distribution of PM2.5 Levels by Month')
plt.xlabel('Month')
plt.ylabel('PM2.5 Levels (µg/m³)')
plt.show()
```

```python
# Monthly Trends in Average PM2.5 Levels
monthly_avg = data.groupby(['year', 'month'])['pm2.5'].mean().unstack()

plt.figure(figsize=(12, 8))
monthly_avg.plot()
plt.title('Average PM2.5 Levels by Month Over Years')
plt.xlabel('Year')
plt
```
